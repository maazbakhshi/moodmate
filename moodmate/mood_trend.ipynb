```python
import json
import matplotlib.pyplot as plt
from datetime import datetime

# Load journal log from file
with open("log.json", "r") as f:  
    logs = json.load(f)

print(f"Loaded {len(logs)} journal entries.")
```

```python
# Prepare mood mapping
mood_map = {
   "happy": 1,
    "happy üòä": 1,
    "neutral": 0,
    "neutral üòê": 0,
    "sad": -1,
    "sad üò¢": -1
}

# Extract dates and mood scores
dates = []
mood_scores = []

for entry in logs:
    # Parse the timestamp string to a datetime object
    dt = datetime.strptime(entry["timestamp"], "%Y-%m-%d %H:%M")
    dates.append(dt)
    
    mood_label = entry["mood"]
    mood_score = mood_map.get(mood_label, 0)  # default to neutral if unknown
    mood_scores.append(mood_score)
```

```python
# Plot the mood trend
plt.figure(figsize=(10, 5))
plt.plot(dates, mood_scores, marker="o", linestyle="-", color="purple")

# Label the graph
plt.title("üß† Mood Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Mood Level")
plt.yticks([-1, 0, 1], ["Sad üò¢", "Neutral üòê", "Happy üòä"])
plt.grid(True)

plt.tight_layout()
plt.show()
```
